```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
data = np.genfromtxt('ccpp_x_y_train.csv', delimiter=',')

X_train = data[:, 0:4]
Y_train = data[:, 4]

X_test = np.genfromtxt('ccpp_x_test.csv', delimiter=',')

# Feature scaling - since data is provided without normalization
scaler = preprocessing.StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

# add self 2 deg features
for i in range(X_train.shape[1]):
    new_column = (X_train[:, i] * X_train[:, i]).reshape(-1,1)
    X_train = np.append(X_train, new_column, axis=1)

for i in range(X_test.shape[1]):
    new_column = (X_test[:, i] * X_test[:, i]).reshape(-1,1)
    X_test = np.append(X_test, new_column, axis=1)
    
## Add row of ones to testing data
train_ones = np.ones(X_train.shape[0]).reshape(-1, 1)
X_train = np.append(X_train, train_ones, axis=1)
## Add row of ones to training data
test_ones = np.ones(X_test.shape[0]).reshape(-1, 1)
X_test = np.append(X_test, test_ones, axis=1)

X_train.shape, Y_train.shape, X_test.shape, train_ones.shape, test_ones.shape
```

### Gradient Descent N Features

```python
def step_gradient(X_train, Y_train, learning_rate, m, j):
    m_j = 0
    data_pts = X_train.shape[0]
    M = len(m)
    for i in range(data_pts):
        # for all data points
        x_i = X_train[i, :]
        y_i = Y_train[i]
        temp_sum = 0
        for k in range(M):
            temp_sum += m[k]*x_i[k]
        temp_sum = y_i - temp_sum
        m_j += (-2/data_pts) * (temp_sum) * x_i[j]
    m[j] = m[j] - (learning_rate * m_j)
    return m[j]
```

```python
def gradient_descent(X_train, Y_train, learning_rate, num_iterations):
    m = [0]*(X_train.shape[1])
    m[-1] = 1 #c
    M = len(m)
    for i in range(num_iterations):
        for j in range(M):
            m[j] = step_gradient(X_train, Y_train, learning_rate, m, j)
        print('Cost - ', i, cost(X_train, Y_train, m))
    return m
```

```python
def cost(X_train, Y_train, m):
    cost = 0
    data_pts = len(X_train)
    N = len(m)
    for i in range(data_pts):
        x_i = X_train[i, :]
        y_i = Y_train[i]
        temp_sum = 0
        for k in range(N):
            temp_sum += m[k]*x_i[k]
        temp_sum = y_i - temp_sum
        cost += (1/data_pts) * ((temp_sum)**2)
    return cost
```

```python
def predict(X_test, m):
    n_fts = X_test.shape[1]
    n_m = np.array(m).reshape(n_fts, 1)
    return np.dot(X_test, n_m)
```

```python
def run():
    # Initialize parameters for GD
    num_iterations = 1200
    learning_rate = 0.1
    m = gradient_descent(X_train, Y_train, learning_rate, num_iterations)
    print(m)
    Y_pred = predict(X_test, m)
    print(Y_pred)
    # save preds to csv
    np.savetxt('pred.csv', Y_pred, fmt='%.5f', delimiter=',')
```

```python
run()
```
